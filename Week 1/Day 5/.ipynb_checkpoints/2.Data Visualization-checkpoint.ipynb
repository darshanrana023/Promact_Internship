```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

dt=pd.read_csv('organizations-100.csv')
```

```python
dt.head()
```

```python
sns.histplot(x='Number of employees',data=dt)
plt.show()
```

```python
sns.boxplot(x='Number of employees', y='Founded',data=dt)
```

```python
sns.scatterplot(x='Country',y='Name', data=dt ,hue='Index', size='Number of employee')

# Placing Legend outside the Figure
plt.legend(bbox_to_anchor=(1, 1))

plt.show()
```
