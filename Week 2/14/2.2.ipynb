# Create pair plots, violin plots, and box plots for a dataset. 

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
sns.get_dataset_names()
```

```python
dim=sns.load_dataset('diamonds')
dim
```

# Pair Plot

```python
sns.pairplot(data=dim,vars=["carat", "depth", "table", "price"])
```


# Violin Plot

```python
sns.violinplot(data=dim['price'])
sns.violinplot(data=dim['carat'])
```

```python
sns.violinplot(data=dim['depth'])
```

```python
dim
```

# Box Plot

```python
sns.boxplot(data=dim,x='cut',y='price')
```

```python
max(dim['price'])
```

```python
sns.heatmap(dim.corr())
plt.title('Correlation Heatmap of Diamond Features')
plt.show()
```
