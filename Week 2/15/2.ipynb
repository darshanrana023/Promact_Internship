3.Group data, aggregate information, and derive insights from a dataset

```python
new.head()
```

```python
grouped_data=new.groupby("Subscription").agg(
    Total_Users=("User ID", "count"),
  
    Average_Amount=("Monthly Revenue", "mean")
)
print("Grouped and Aggregated Data:")
print(grouped_data)
```

```python
monthly_revenue=new['Monthly Revenue'].sum()
print('Total Revenue:',monthly_revenue)
```

```python
average_age_of_viewer= new["Age"].mean()
print("Average Age of viewers:",average_age_of_viewer)
average_monthly_revenue= new["Monthly Revenue"].mean()
print("Average Monthly Revenue:",average_monthly_revenue)
```
