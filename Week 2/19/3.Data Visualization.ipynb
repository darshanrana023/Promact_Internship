```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset

data = pd.read_csv('Admission.csv')

def visualize_histogram(column_data):
    plt.figure(figsize=(8, 6))
    plt.hist(column_data, bins=20, edgecolor='k', alpha=0.7)
    plt.xlabel('Value')
    plt.ylabel('Frequency')
    plt.title('Histogram')
    plt.show()

def visualize_scatter(x_column, y_column):
    plt.figure(figsize=(8, 6))
    plt.scatter(data[x_column], data[y_column], alpha=0.5)
    plt.xlabel(x_column)
    plt.ylabel(y_column)
    plt.title('Scatter Plot')
    plt.show()

# Allow users to select columns and visualize data distributions using histograms
selected_column = input("Enter the column name for histogram: ")
if selected_column in data.columns:
    visualize_histogram(data[selected_column])
else:
    print("Column not found in the dataset.")

# Implement scatter plots for bivariate data visualization
x_column = input("Enter the x-axis column name for scatter plot: ")
y_column = input("Enter the y-axis column name for scatter plot: ")
if x_column in data.columns and y_column in data.columns:
    visualize_scatter(x_column, y_column)
else:
    print("One or both columns not found in the dataset.")
```
