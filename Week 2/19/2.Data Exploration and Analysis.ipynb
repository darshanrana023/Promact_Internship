```python
import pandas as pd

def clean_data(data, fill_mode='specified', fill_value=None, drop_threshold=0.5):
    # Calculate missing values for each column
    missing_values = data.isnull().sum()
    
    # Display columns with missing values
    print("Columns with missing values:")
    print(missing_values[missing_values > 0])
    
    # Filling missing values
    if fill_mode == 'specified':
        if fill_value is not None:
            data.fillna(fill_value, inplace=True)
    elif fill_mode == 'mean':
        data.fillna(data.mean(), inplace=True)
    elif fill_mode == 'median':
        data.fillna(data.median(), inplace=True)
    
    # Drop columns with high percentage of missing values
    drop_columns = missing_values[missing_values / len(data) > drop_threshold].index
    data.drop(columns=drop_columns, inplace=True)
    
    return data


data = pd.read_csv('Salary.csv')

# User choices
fill_mode_choice = input("Choose fill mode (specified/mean/median): ")
if fill_mode_choice == 'specified':
    fill_value_choice = input("Enter fill value: ")
    fill_value = float(fill_value_choice)  # Convert to appropriate data type if needed
else:
    fill_value = None

drop_threshold_choice = float(input("Enter drop threshold (0-1): "))

# Clean the data
cleaned_data = clean_data(data, fill_mode=fill_mode_choice, fill_value=fill_value, drop_threshold=drop_threshold_choice)

# Display cleaned dataset
print("Cleaned dataset:")
print(cleaned_data)
```
