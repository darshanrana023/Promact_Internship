```python
import pandas as pd
import numpy as np

# Replace with the actual path to the uploaded CSV file

# Load the dataset into a DataFrame
data = pd.read_csv('reviews.csv')

# Calculate and display basic statistics
statistics = {
    'Column': [],
    'Mean': [],
    'Median': [],
    'Std Deviation': [],
    'Missing Values': []
}

for column in data.columns:
    mean = data[column].mean()
    median = data[column].median()
    std_dev = data[column].std()
    missing_values = data[column].isnull().sum()
    
    statistics['Column'].append(column)
    statistics['Mean'].append(mean)
    statistics['Median'].append(median)
    statistics['Std Deviation'].append(std_dev)
    statistics['Missing Values'].append(missing_values)

statistics_df = pd.DataFrame(statistics)
print(statistics_df)
```
